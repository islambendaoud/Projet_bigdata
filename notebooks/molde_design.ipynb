```python
import pandas as pd 
```

```python
df_u = pd.read_csv("usagers_file.csv" , sep = ';' , encoding= "utf-8")
df_v = pd.read_csv("vehicule_file.csv" , sep = ';')
df_l = pd.read_csv("lieux_file.csv", sep = ';')
```

```python
df_u.head()
```

```python
df_v.head()
```

```python
df_l.head()
```

```python
# join usagers and vehicule on id_vehicle and delete the things that are repeated instead of giving x and y to the name 

df_uv = pd.merge(df_u, df_v, on='id_vehicule' , suffixes=('','_y'))
df_uv = df_uv.drop(df_uv.filter(regex='_y$').columns.tolist(),axis=1)
df = pd.merge(df_uv, df_l, on='Num_Acc' , suffixes=('','_y'))
df = df.drop(df.filter(regex='_y$').columns.tolist(),axis=1)
```

```python
# count na of every column
df.isna().sum()
```

```python
# drop the columns that have more than 50% of na
df.dropna(thresh=0.5*len(df), axis=1, inplace=True)
```

```python
# drop the columns that are not float or int 
df = df.select_dtypes(exclude=['object'])
```

```python
df.head()
```

```python
df.drop(columns =[ 'Num_Acc'], inplace=True)
```

```python
df.head()
```

```python
import numpy as np
```

```python
# import train_test_split from sklearn
from sklearn.model_selection import train_test_split

# y is df['grav'] and X is df.drop(columns=['grav'])
y = df['grav']
X = df.drop(columns=['grav'])


# treat the na values by replacing them with the mean of the column
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(X)
X = imp.transform(X)


# split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)




# import RandomForestClassifier from sklearn
from sklearn.ensemble import RandomForestClassifier

# create a RandomForestClassifier model
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# put 0 where -1 in y_train and y_test
y_train = y_train.replace(-1,0)
y_test = y_test.replace(-1,0)
```

```python
# xgboost
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# hyperopt on xgboost
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
from sklearn.metrics import accuracy_score

# define the search space
# too big , it takes too much time to run give me smaller space
space = {
    'max_depth': hp.choice('max_depth', [5, 10, 15, 20, 25, 30]),
    'n_estimators': hp.choice('n_estimators', [100, 200, 300 , 600 ]),
    'learning_rate': hp.uniform('learning_rate', 0.001, 0.1),
    'subsample': hp.uniform('subsample', 0.5, 1),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1),
    'gamma': hp.uniform('gamma', 0, 1),
}

# define the objective function
def objective(params):
    params = {
        'max_depth': int(params['max_depth']),
        'n_estimators': int(params['n_estimators']),
        'learning_rate': params['learning_rate'],
        'gamma': params['gamma'],
        'colsample_bytree': params['colsample_bytree'],
        'subsample': params['subsample'],
    }
    
    model = XGBClassifier(**params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return 1 - score 

# run the hyper parameter search
trials = Trials()
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=50,
            trials=trials)
```
